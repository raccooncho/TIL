
# How to Reset DB

1. 특정 App과 관련된 모든 DB table 을 DROP 한다.

```bash
$ python manage.py migrate [APP_NAME(otoo)] zero
```

2. 해당 App의 migrations/ 안의 마이그레이션 파일을 삭제한다.

```bash
$ rm [APP_NAME](otoo)/migrations/0*
```

3. 다시 마이그레이션 파일을 만든다. (당연히 0001_initial.py 만 나오는게 정상)

```bash
$ python manage.py makemigrations [APP_NAME(otoo)]
```

4. 다시 마이그레이트

```bash
$ python manage.py migrate
```
